```python
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

# Divide subjects into train, val and test in the same order as axial slices

```python
def load_cross_validation_settings(cv_file):
    cv_setting = None
    with open(cv_file, "rb") as obj:
        cv_setting = pickle.load(obj)
        
    return cv_setting
```

```python
data_dict = load_cross_validation_settings('Alz_MCI_vs_AD_1365_each_full_subject_list.pkl')
```

```python
data_dict.keys()
```

```python
train = data_dict['train']
validation = data_dict['validation']
test = data_dict['test']
```

```python
train.keys()
```

```python
subject_ids_train = train['subject_dict']
subject_ids_val = validation['subject_dict']
subject_ids_test = test['subject_dict']
```

```python
subject_group_train = train['subject_group']
subject_group_val = validation['subject_group']
subject_group_test = test['subject_group']
```

```python
subject_ids_train
```

```python
slice_range = range(18,130)
```

```python
subject_slices_train = {}
subject_slices_val = {}
subject_slices_test = {}

for i in list(subject_ids_train.keys()):
    slice_list = []
    for s in slice_range:
        slice_name = i + '_' + str(s) + '.npy'
        slice_list.append(slice_name)
        
    subject_slices_train[i] = slice_list
    
for i in list(subject_ids_val.keys()):
    slice_list = []
    for s in slice_range:
        slice_name = i + '_' + str(s) + '.npy'
        slice_list.append(slice_name)
        
    subject_slices_val[i] = slice_list
    
for i in list(subject_ids_test.keys()):  
    slice_list = []
    for s in slice_range:
        slice_name = i + '_' + str(s) + '.npy'
        slice_list.append(slice_name)
        
    subject_slices_test[i] = slice_list
```

```python
coronal_data_dict = {}
coronal_data_dict['train'] = {'subject_dict':train['subject_dict'], 'subject_group':train['subject_group'],'subject_slices':subject_slices_train}
coronal_data_dict['validation'] = {'subject_dict':validation['subject_dict'], 'subject_group':validation['subject_group'],'subject_slices':subject_slices_val}
coronal_data_dict['test'] = {'subject_dict':test['subject_dict'], 'subject_group':test['subject_group'],'subject_slices':subject_slices_test}
```

```python
f = open('Alz_MCI_vs_AD_1365_each_full_subject_list_coronal.pkl','wb')
pickle.dump(coronal_data_dict, f)
f.close()
```
